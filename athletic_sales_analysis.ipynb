```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv("athletic_sales_2020.csv")
sales_2021 = pd.read_csv("athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
display(sales_2020)
```

```python
# Display the 2021 sales DataFrame
display(sales_2021)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

combined_sales = pd.concat([sales_2020, sales_2021], join = 'inner', axis ='rows')
combined_sales.reset_index(inplace=True)
combined_sales
```

```python
# Check if any values are null.
combined_sales.isnull()
```

```python
# Check the data type of each column
combined_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales["invoice_date"] = pd.to_datetime(combined_sales["invoice_date"])

```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
region_state_city_df = combined_sales.groupby(["region", "state", "city"])[["units_sold"]].sum()
region_state_city_df = region_state_city_df.rename(columns={"units_sold" : "Total_Products_Sold"})
rsc_sorted = region_state_city_df.sort_values(by="Total_Products_Sold", ascending = False)

# Show the top 5 results.
rsc_sorted.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_sales_pivot = pd.pivot_table(combined_sales,
                                   values = "total_sales",
                                   index = ["region", "state", "city"],
                                   aggfunc= "sum")

# Rename the "units_sold" column to "Total_Products_Sold"
total_sales_pivot = total_sales_pivot.rename(columns={"total_sales" : "Total Sales"})
total_sales_pivot_sorted = total_sales_pivot.sort_values(by="Total Sales", ascending = False)

# Show the top 5 results.
total_sales_pivot_sorted.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = combined_sales.groupby(['region','state','city'])[['total_sales']].sum()
total_sales = total_sales.rename(columns={'total_sales':'Total Sales'})
total_sales_sort = total_sales.sort_values(by = 'Total Sales', ascending = False)

# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot = pd.pivot_table(combined_sales,
                                   values = 'total_sales',
                                   index = ['region','state','city'],
                                   aggfunc= 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot = total_sales_pivot.rename(columns={'total_sales':'Total Sales'})
total_sales_pivot_sort= total_sales_pivot.sort_values(by = 'Total Sales', ascending = False)

# Show the top 5 results.
total_sales_pivot_sort.head()

```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = combined_sales.groupby(['retailer','region','state','city']).sum('total_sales')
total_sales = total_sales.rename(columns={"total_sales" : "Total Sales"})

total_sales = total_sales.drop(columns= ['retailer_id','price_per_unit','units_sold','operating_profit'])
total_sales_sort = total_sales.sort_values(by=['Total Sales'],ascending=False)

# Show the top 5 results.

total_sales_sort.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales_pivot= combined_sales.pivot_table(index=['retailer','region','state','city'],values='total_sales',aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot = total_sales_pivot.rename(columns={"total_sales" : "Total Sales"})


# Show the top 5 results.
total_sales_pivot = total_sales_pivot.sort_values(by=['Total Sales'], ascending=False)
total_sales_pivot.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear = combined_sales.loc[(combined_sales['product'] == 'Women\'s Athletic Footwear')]
womens_footwear.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_women = womens_footwear.groupby(['retailer','region','state','city']).sum('units_sold')
total_women = total_women.rename(columns={"units_sold" : "Women Footwear Units Sold"})

total_women = total_women.drop(columns= ['retailer_id','price_per_unit','total_sales','operating_profit'])
total_women_sort = total_women.sort_values(by=['Women Footwear Units Sold'],ascending=False)

# Show the top 5 results.
total_women_sort.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
total_women_pivot = pd.pivot_table(womens_footwear,
                                   values = 'units_sold',
                                   index = ['retailer','region','state','city'],
                                   aggfunc= 'sum')


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_women_pivot = total_women_pivot.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})
total_women_pivot_sort = total_women_pivot.sort_values(by = 'Womens_Footwear_Units_Sold', ascending = False)

# Show the top 5 results.
total_women_sort.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_pivot = pd.pivot_table(womens_footwear,
                                   values = 'total_sales',
                                   index = 'invoice_date',
                                   aggfunc= 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_pivot = womens_pivot.rename(columns={'total_sales':'Total Sales'})
womens_pivot_sort = womens_pivot.sort_values(by = 'Total Sales', ascending = False)

# Show the table.
womens_pivot_sort.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
resampled_pivot = womens_pivot.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
resampled_pivot_sort = resampled_pivot.sort_values(by='Total Sales', ascending = False)
resampled_pivot_sort.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
resampled_pivot_week = womens_pivot.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
resampled_pivot_week_sort = resampled_pivot_week.sort_values(by='Total Sales', ascending = False)
resampled_pivot_week_sort.head(10)
```
